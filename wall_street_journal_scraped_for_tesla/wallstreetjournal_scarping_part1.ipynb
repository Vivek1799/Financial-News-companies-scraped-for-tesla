```python
import re

def extract_info(text):
    headline = re.search("Headline: (.*?)\n", text)
    excerpt = re.search("Excerpt: (.*?)\n", text)
    timestamp = re.search("Timestamp: (.*?)\n", text)
    
    if headline:
        headline = headline.group(1)
    else:
        headline = None
        
    if excerpt:
        excerpt = excerpt.group(1)
    else:
        excerpt = None
        
    if timestamp:
        timestamp = timestamp.group(1)
    else:
        timestamp = None
    
    return (headline, excerpt, timestamp)

text = "Headline: Example Headline\nExcerpt: Example Excerpt\nTimestamp: 2021-01-01 12:00:00\n"
print(extract_info(text))
```
