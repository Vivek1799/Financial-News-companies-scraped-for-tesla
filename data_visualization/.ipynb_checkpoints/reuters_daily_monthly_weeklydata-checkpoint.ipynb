```python
import pandas as pd

# Assuming your df_reuters_count has columns Date and Count
df_monthly = df_reuters_count.copy()

# Shift the dates to start from 10th day of each month
df_monthly['Month'] = (df_monthly['Date'] + pd.offsets.Day(9)).dt.to_period('M')

# Group by the shifted dates and sum the counts
df_monthly = df_monthly.groupby('Month')['Count'].sum().reset_index()
```

```python
# Filter out the first and last weeks of each month
df_filtered = df_reuters_count[(df_reuters_count['Date'].dt.day >= 10) & (df_reuters_count['Date'].dt.is_month_end == False)]

# Group by month and sum the counts
df_monthly = df_filtered.groupby(pd.Grouper(key='Date', freq='MS')).sum().reset_index()

# Rename the Date column to Month
df_monthly = df_monthly.rename(columns={'Date': 'Month'})
```
