```python
import pandas as pd
```

```python
df_ft = pd.read_csv("ft_cleaned.csv")
```

```python
df_ft
```

```python
df_ft_count = df_ft.groupby('Timestamps')['Headlines'].count().reset_index()
df_ft_count = df_ft_count.rename(columns={'Timestamps': 'Date', 'Headlines': 'Count'})
```

```python
df_ft_count
```

```python
df_ft_count.to_csv("ft_daily.csv")
```

```python
max_count = df_ft_count['Count'].max()
min_count = df_ft_count['Count'].min()

print("Maximum count:", max_count)
print("Minimum count:", min_count)
```

```python
# Sort by count in descending order to get the maximum count
df_ft_count_sorted_desc = df_ft_count.sort_values(by='Count', ascending=False)
max_date_count = df_ft_count_sorted_desc.head(1)

# Sort by count in ascending order to get the minimum count
df_ft_count_sorted_asc = df_ft_count.sort_values(by='Count', ascending=True)
min_date_count = df_ft_count_sorted_asc.head(1)

print("Date with maximum count:")
print(max_date_count)

print("Date with minimum count:")
print(min_date_count)
```

```python
import pandas as pd

df_ft_count['Date'] = pd.to_datetime(df_ft_count['Date'])
```

```python
df_ft_count.dtypes
```

```python
df_ft_count_weekly = df_ft_count.resample('W', on='Date').count().reset_index()
```

```python
df_ft_count_weekly = df_ft.resample('W', on='Timestamps').count().reset_index()
```

```python
df_ft_count_weekly
```

```python
import pandas as pd

# Assuming the `Date` column is in datetime format
df_ft_count_weekly['Week'] = df_ft_count_weekly['Date'].apply(lambda x: x.strftime('%Y, Week %U'))

# Display the updated DataFrame
print(df_ft_count_weekly)
```

```python
df_ft_count_weekly
```

```python
df_ft_count_weekly.to_csv("ft_weekly.csv")
```

```python
import pandas as pd

# Assuming your Date column is already in datetime format
df_monthly = df_ft_count.set_index('Date').resample('MS').sum()

# Shift the dates to start from 10th day of each month
df_monthly.index = df_monthly.index + pd.DateOffset(days=9)

# Reset the index and rename the columns
df_monthly = df_monthly.reset_index().rename(columns={'Date': 'Month', 'Count': 'Monthly Count'})
```

```python
df_monthly
```

```python
df_monthly.to_csv("ft_monthly.csv")
```

```python
df_ft_count = df_ft.groupby(pd.Grouper(key='Timestamps', freq='D')).size().reset_index(name='Count')
```

```python
# Select headlines for Nov-Dec period
nov_dec_count = df_ft.query("'2022-11-10' <= Timestamps <= '2022-12-09'")['Headlines'].count()

# Select headlines for Dec-Jan period
dec_jan_count = df_ft.query("'2022-12-11' <= Timestamps <= '2023-01-10'")['Headlines'].count()

# Select headlines for Jan-Feb period
jan_feb_count = df_ft.query("'2023-01-11' <= Timestamps <= '2023-02-10'")['Headlines'].count()

# Create a dictionary to store the results
data = {'Month': ['Nov-Dec', 'Dec-Jan', 'Jan-Feb'], 'Count': [nov_dec_count, dec_jan_count, jan_feb_count]}

# Create the DataFrame
df_monthly = pd.DataFrame(data)
```

```python
df_monthly.to_csv("ft_monthly_revised.csv")
```
