```python
import pandas as pd
```

```python
df_wsj = pd.read_csv("wsj_cleaned.csv")
```

```python
df_wsj_count = df_wsj.groupby('Timestamps')['Headlines'].count().reset_index()
df_wsj_count = df_wsj_count.rename(columns={'Timestamps': 'Date', 'Headlines': 'Count'})
```

```python
df_wsj_count
```

```python
df_wsj_count.to_csv("wsj_daily.csv")
```

```python
max_count = df_wsj_count['Count'].max()
min_count = df_wsj_count['Count'].min()

print("Maximum count:", max_count)
print("Minimum count:", min_count)
```

```python
# Sort by count in descending order to get the maximum count
df_wsj_count_sorted_desc = df_wsj_count.sort_values(by='Count', ascending=False)
max_date_count = df_wsj_count_sorted_desc.head(1)

# Sort by count in ascending order to get the minimum count
df_wsj_count_sorted_asc = df_wsj_count.sort_values(by='Count', ascending=True)
min_date_count = df_wsj_count_sorted_asc.head(1)

print("Date with maximum count:")
print(max_date_count)

print("Date with minimum count:")
print(min_date_count)
```

```python
import pandas as pd

df_wsj_count['Date'] = pd.to_datetime(df_wsj_count['Date'])
```

```python
df_wsj_count_weekly = df_wsj_count.resample('W', on='Date').count().reset_index()
```

```python
df_wsj_count_weekly
```

```python
import pandas as pd

# Assuming the `Date` column is in datetime format
df_wsj_count_weekly['Week'] = df_wsj_count_weekly['Date'].apply(lambda x: x.strftime('%Y, Week %U'))

# Display the updated DataFrame
print(df_wsj_count_weekly)
```

```python
df_wsj_count_weekly
```

```python
df_wsj_count_weekly.to_csv("wsj_weekly.csv")
```

```python
df_wsj_count_monthly = df_wsj_count.resample('M', on='Date').count()
```

```python
df_wsj_count_monthly
```

```python
import pandas as pd

# Assuming your Date column is already in datetime format
df_monthly = df_wsj_count.set_index('Date').resample('MS').sum()

# Shift the dates to start from 10th day of each month
df_monthly.index = df_monthly.index + pd.DateOffset(days=9)

# Reset the index and rename the columns
df_monthly = df_monthly.reset_index().rename(columns={'Date': 'Month', 'Count': 'Monthly Count'})
```

```python
df_monthly
```

```python
df_monthly.to_csv("wsj.monthly.csv")
```
