# function to scrape all the data

```python
import asyncio
import time

from playwright.sync_api import sync_playwright

async def scrape_headlines_and_timestamps(page):
    headlines = []
    excerpts = []
    timestamps = []
    
    element_count = await page.evaluate("document.querySelectorAll('div.o-teaser__heading').length")
    for i in range(0, element_count):
        headline = await page.locator("div.o-teaser__heading").nth(i).inner_text()
        headlines.append(headline)

    
    element_count = await page.evaluate("document.querySelectorAll('p.o-teaser__standfirst').length")
    for i in range(0, element_count):
        excerpt = await page.locator("p.o-teaser__standfirst").nth(i).inner_text()
        excerpts.append(excerpt)
    
    element_count = await page.evaluate("document.querySelectorAll('time.o-teaser__timestamp-date').length")
    for i in range(0, element_count):
        timestamp = await page.locator("time.o-teaser__timestamp-date").nth(i).inner_text()
        timestamps.append(timestamp)

        
    return headlines, excerpts, timestamps

async def main():
    async with sync_playwright() as playwright:
        browser = playwright.chromium
# Create a new page
        page = await browser.new_page()
        await scrape_headlines_and_timestamps(page)
        await asyncio.sleep(4)

all_headlines = []
all_excerpts = []
all_timestamps = []          
while True:
            headlines, excerpts, timestamps = await scrape_headlines_and_timestamps(page)
            all_headlines.extend(headlines)
            all_excerpts.extend(excerpts)
            all_timestamps.extend(timestamps)
#             print(headlines)
#             print(timestamps)
            try:
                await page.locator("xpath=/html/body/div/div[2]/div/div/div/main/div/div[3]/div/a[2]").click()
                await asyncio.sleep(4)
            except:
                break
#             print(all_headlines)
#             print(all_timestamps)    

async def main():
    async with sync_playwright() as playwright:
        # Start a browser
        browser = playwright.chromium
        # Create a new page
        page = await browser.new_page()
        await scrape_headlines_and_timestamps(page)
        await asyncio.sleep(4)

print(all_headlines)
print(all_excerpts)
print(all_timestamps)
```
