```python
import pandas as pd
```

```python
df = pd.read_csv("reuters_scraped_timestamp.csv")
```

```python
df
```

```python
df['time_stamps'] = pd.to_datetime(df['time_stamps'], format='%m/%d/%Y')
```

```python
df
```

```python
df.dtypes
```

```python
df.shape
```

```python
start_date = pd.Timestamp('2022-11-10')
end_date = pd.Timestamp('2023-02-10')

# df['time_stamps'] = pd.to_datetime(df['time_stamps'], format='%m/%d/%Y')

# print(df['time_stamps'].dtype) # check if time_stamps has been converted to datetime correctly

mask = (df['time_stamps'] >= start_date) & (df['time_stamps'] <= end_date)
df = df.loc[mask]

print(df.shape) # check if the number of rows has changed after applying the filter

df = df[['headlines', 'all_content', 'time_stamps']]
```

```python
df.head(30000)
```

```python
duplicates = df[df.duplicated()]
num_duplicates = len(duplicates)
print("Number of duplicate values: ", num_duplicates)
```

```python
duplicates = df[df.duplicated(subset=['headlines'], keep=False)]
num_duplicates = len(duplicates)
print("Number of duplicate rows:", num_duplicates)
```
